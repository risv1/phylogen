```python
import json
import pandas as pd
```

```python
CARD_CIPRO_ECOLI_MUTATIONS_PATH = "cipro_ecoli_mutations.json"
```

```python
with open(CARD_CIPRO_ECOLI_MUTATIONS_PATH, 'r') as f:
    card_data = json.load(f)
```

```python
card_df = pd.DataFrame(card_data)
print(f"Loaded {len(card_df)} CARD entries.")
card_df.head()
```

```python
if 'snps' in card_df.columns:
    snps_df = card_df.explode('snps').reset_index(drop=True)
    # Unpack snps dict into columns
    snps_df = pd.concat([snps_df.drop(['snps'], axis=1), snps_df['snps'].apply(pd.Series)], axis=1)
    if 'protein_sequence' in snps_df.columns:
        snps_df = snps_df.drop(columns=['protein_sequence'])
    
    snps_df.to_csv("cipro_snps_df.csv", index=False)
    print("Saved flattened SNPs DF to 'cipro_snps_df.csv'.")
else:
    print("No 'snps' column found—check JSON structure.")
```
